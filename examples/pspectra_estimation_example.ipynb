### How the Script Works
The script works by taking a file containg epoch data, and transforms the data into UVP objects. The UVP object will contain all the powerspectra created from the epoch and its corresponding metadata. From the UVP object, keys will be attained that are a tuple in the following format (spectral window, baseline-pair, polarization-string). The keys will then be sorted out, where any keys that have the flag sum of 0 or baseline pairs, or where the antennas are the same will be removed. Once the keys have been properly sorted, the sidereal time, highdelays, average power, and standard deviation will be generated and plotted accordingly.
